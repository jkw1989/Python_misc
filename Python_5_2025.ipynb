# Matplotlib

```python
import pandas as pd
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = df['job_posted_date']
y = df['job_posted_date'].size
```

```python
monthly_job_counts = df.set_index('job_posted_date').resample('ME').size()
x = monthly_job_counts.index
y = monthly_job_counts.values
plt.plot(x, y)
plt.xlabel('Date (Monthly)')
plt.ylabel('Number of Job Postings')
plt.title('Monthly Job Postings Over Time')
plt.show()
```

```python
months = ['January', 'February', 'March', 'April']
postings = [120, 150, 100, 130]

plt.plot(months, postings)
plt.show()
```

```python
job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']
postings = [50, 75, 60]

plt.bar(job_titles,postings)
plt.show()
```

```python
counts = df['job_title_short'].value_counts()#head(3)

plt.bar(counts.index, counts.values)
plt.ylabel('Number of Postings')
plt.title('Job Postings by Title - Top 3')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
plt.plot(df.job_posted_date,df.job_posted_date)
```

```python
job_month = df['job_posted_date'].dt.month

# Display the first few rows with the new column
job_month.head()
```

```python
monthly_counts = job_month.value_counts().sort_index()

plt.bar(monthly_counts.index, monthly_counts.values)
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Job Postings by Month')
plt.xticks(range(1, 13))
plt.show()
```

```python
job_week = df['job_posted_date'].dt.day.sort_index()

weekly_counts = job_week.value_counts().sort_index()

plt.bar(weekly_counts.index, weekly_counts.values)
plt.ylabel('Number of Job Postings')
plt.title('Job Postings by Month')
#plt.xticks(range(0,31,10))
plt.show()
```

```python
job_salaries = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
job_salaries

plt.barh(job_salaries.index,job_salaries.values)
```

```python
job_month = df['job_posted_date'].dt.month

job_monthly_salary = df.groupby(job_month)['salary_year_avg'].mean()

plt.plot(job_monthly_salary.index,job_monthly_salary.values,color="red",label="salary")
plt.title("This title")
plt.xlabel("Month")
plt.ylabel("Average Salary")
plt.legend()
plt.show()
```

```python
country_count = df['job_country'].value_counts().head(8)

plt.bar(country_count.index, country_count.values)
plt.title("Count by country")
plt.xlabel("Coutries")
plt.ylabel("Postings")
plt.xticks(rotation=45, ha='right')

plt.show()
```
